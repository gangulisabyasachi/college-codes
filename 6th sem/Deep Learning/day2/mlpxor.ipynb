```python
import numpy as np
```

```python
def relu(x):
    if x>=0:
        return x
    return 0
```

```python
X = np.array([
    [0,0],
    [0,1],
    [1,0],
    [1,1]
])

Y = np.array([0,1,1,0])
```

```python
W1 = np.array([
    [1,-1],
    [-1,1]
])
b1 = np.array([-0.5,-0.5])
W2 = np.array([1,1])
b2 = np.array([-0.5])
```

```python
def activ(x):
    return x>=0
```

```python
z1 = X @ W1.T + b1
h1 = activ(z1)
```

```python
z2 = h1 @ W2.T + b2
h2 = activ(z2)
```
