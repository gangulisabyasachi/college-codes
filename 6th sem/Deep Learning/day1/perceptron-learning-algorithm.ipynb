## AND GATE

```python
X=np.array([(0,0),(0,1),(1,0),(1,1)])
Y=np.array([0,0,0,1])
```

```python
W=np.zeros(2)
b = -0.5
```

```python
lr = 1
```

```python
def activ(x):
    return x>=0
```

```python
W.shape, X.shape
```

```python
EPOCH=10
for _ in range(EPOCH):
    for i in range(len(X)):
        z=np.dot(X[i],W)+b
        y_pred = activ(z)
        err = Y[i] - y_pred
        W = W + lr*err*X[i]
        b=b+lr*err


print(W)
print(b)
```
