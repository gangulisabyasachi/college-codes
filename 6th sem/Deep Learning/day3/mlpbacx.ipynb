```python
import numpy as np
```

```python
X = np.array([
    [63,45,233],
    [41,130,204],
    [37,130,250],
    [56,120,236],
    [57,120,236]
])
y = np.array([[1],[1],[1],[0],[0]])
```

```python
X = X/np.max(X,axis = 0)
np.random.seed(1)
W1 = np.random.rand(3,5)
b1 = np.zeros((1,5))
W2 = np.random.rand(5,1)
b2 = np.zeros((1,1))
```

```python
lr = 0.5
epoches = 1000
```

```python
def sigmoid(z):
    return 1 / (1 + np.exp(-z))
```

```python
for i in range(epoches):

    z1 = np.dot(X, W1) + b1
    a1 = sigmoid(z1)

    z2 = np.dot(a1, W2) + b2
    output = sigmoid(z2)

    err = y - output
    d_out = err * output * (1 - output)

    d_hidden = np.dot(d_out, W2.T) * a1 * (1 - a1)

    W2 += lr * np.dot(a1.T, d_out)
    b2 += lr * np.sum(d_out, axis=0, keepdims=True)

    W1 += lr * np.dot(X.T, d_hidden)
    b1 += lr * np.sum(d_hidden, axis=0, keepdims=True)

print(np.round(output,3))
```

```python
print((output>0.5).astype(int))
```
