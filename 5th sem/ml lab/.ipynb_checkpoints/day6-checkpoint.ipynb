```python
import pandas as pd

df = pd.read_csv('/Users/sabyasachiganguli/Desktop/code/college codes/5th sem/ml lab/day6/train.csv')
```

```python
display(df.head())
```

```python
import matplotlib.pyplot as plt

independent_variables = ['GrLivArea', 'TotalBsmtSF', '1stFlrSF', 'GarageCars', 'GarageArea']

for var in independent_variables:
    plt.figure(figsize=(8, 6))
    plt.scatter(df[var], df['SalePrice'])
    plt.title(f'Scatter Plot of SalePrice vs. {var}')
    plt.xlabel(var)
    plt.ylabel('SalePrice')
    plt.show()

# Comment on observed trends and correlations:
print("GrLivArea vs. SalePrice: There appears to be a strong positive linear relationship between GrLivArea and SalePrice. As GrLivArea increases, SalePrice generally increases. There are some outliers with very high GrLivArea but relatively lower SalePrice.")
print("TotalBsmtSF vs. SalePrice: There is a positive linear relationship between TotalBsmtSF and SalePrice. Houses with larger basements tend to have higher sale prices. There are some houses with no basement (TotalBsmtSF = 0) and varying SalePrices.")
print("1stFlrSF vs. SalePrice: Similar to GrLivArea and TotalBsmtSF, there is a positive linear relationship between 1stFlrSF and SalePrice. Larger first floor areas are associated with higher sale prices.")
print("GarageCars vs. SalePrice: There is a positive correlation between GarageCars and SalePrice. Houses with more garage capacity tend to have higher sale prices. The relationship appears somewhat stepped due to the discrete nature of the 'GarageCars' variable.")
print("GarageArea vs. SalePrice: There is a positive linear relationship between GarageArea and SalePrice. Larger garage areas are associated with higher sale prices. This relationship is similar to 'GarageCars' but appears more continuous.")
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt

# 1. Select the 'GrLivArea' column as the independent variable X_simple
X_simple = df[['GrLivArea']]

# 2. Select the 'SalePrice' column as the dependent variable y_simple
y_simple = df['SalePrice']

# 4. Instantiate and fit a linear regression model
model_simple = LinearRegression()
model_simple.fit(X_simple, y_simple)

# 5. Generate predictions
y_simple_pred = model_simple.predict(X_simple)

# 6. Create a scatter plot of X_simple vs. y_simple
plt.figure(figsize=(10, 6))
plt.scatter(X_simple, y_simple, label='Actual Data')

# 7. Plot the regression line
plt.plot(X_simple, y_simple_pred, color='red', linewidth=2, label='Regression Line')
plt.title('Simple Linear Regression: SalePrice vs. GrLivArea')
plt.xlabel('GrLivArea')
plt.ylabel('SalePrice')
plt.legend()
plt.show()

# 8. Print the intercept and coefficient
intercept = model_simple.intercept_
coef = model_simple.coef_[0]
print(f'Intercept: {intercept}')
print(f'Coefficient (Slope) for GrLivArea: {coef}')

# Interpretation of intercept and slope:
print("The intercept represents the predicted SalePrice when GrLivArea is 0. In this context, a GrLivArea of 0 is not meaningful for a house, so the intercept might not have a practical interpretation in isolation, but it's necessary for defining the regression line.")
print("The coefficient (slope) for GrLivArea indicates the average change in SalePrice for a one-unit increase in GrLivArea, assuming a linear relationship. A positive slope suggests that as GrLivArea increases, SalePrice tends to increase.")

# 9. Calculate and print the Mean Squared Error (MSE)
mse_simple = mean_squared_error(y_simple, y_simple_pred)
print(f'Mean Squared Error (MSE): {mse_simple}')

# 10. Calculate and print the R-squared score
r2_simple = r2_score(y_simple, y_simple_pred)
print(f'R-squared Score: {r2_simple}')

# 11. Comment on the performance of the model:
print("The MSE is a measure of the average squared difference between the actual and predicted SalePrices. A lower MSE indicates a better fit. The reported MSE value is relatively large, which is expected given the scale of the SalePrice variable.")
print("The R-squared score represents the proportion of the variance in the dependent variable (SalePrice) that is predictable from the independent variable (GrLivArea). An R-squared of 1 indicates a perfect fit, while 0 indicates no linear relationship. An R-squared of approximately 0.50 suggests that about 50% of the variation in SalePrice can be explained by the linear relationship with GrLivArea. This indicates that while GrLivArea is a significant predictor, a simple linear model based solely on GrLivArea does not fully capture the variability in SalePrice, implying other factors also influence the price.")
```

```python
from sklearn.linear_model import LinearRegression

# 1. Select the specified columns for independent variables X_multiple
X_multiple = df[['GrLivArea', 'TotalBsmtSF', '1stFlrSF', 'GarageCars', 'GarageArea']]

# 2. Select the 'SalePrice' column as the dependent variable y_multiple
y_multiple = df['SalePrice']

# 4. Instantiate a LinearRegression object
model_multiple = LinearRegression()

# 5. Fit the linear regression model
model_multiple.fit(X_multiple, y_multiple)

# 6. Print the intercept
print(f'Intercept: {model_multiple.intercept_}')

# 7. Print the coefficients and their corresponding feature names
print('Coefficients:')
for feature, coef in zip(X_multiple.columns, model_multiple.coef_):
    print(f'{feature}: {coef}')

# 8. Interpret the meaning of each coefficient:
print("Intercept: The intercept represents the predicted SalePrice when all independent variables are zero. In the context of house prices, this value may not have a direct practical interpretation as it's unlikely for all these features to be zero simultaneously.")
print("GrLivArea: The coefficient for GrLivArea indicates the average change in SalePrice for a one-unit increase in above ground living area, holding all other independent variables constant. A positive coefficient suggests that larger living areas are associated with higher sale prices.")
print("TotalBsmtSF: The coefficient for TotalBsmtSF indicates the average change in SalePrice for a one-unit increase in total basement area, holding all other independent variables constant. A positive coefficient suggests that larger basements are associated with higher sale prices.")
print("1stFlrSF: The coefficient for 1stFlrSF indicates the average change in SalePrice for a one-unit increase in first floor area, holding all other independent variables constant. A positive coefficient suggests that larger first floor areas are associated with higher sale prices.")
print("GarageCars: The coefficient for GarageCars indicates the average change in SalePrice for an increase of one car capacity in the garage, holding all other independent variables constant. A positive coefficient suggests that more garage capacity is associated with higher sale prices.")
print("GarageArea: The coefficient for GarageArea indicates the average change in SalePrice for a one-unit increase in garage area, holding all other independent variables constant. A positive coefficient suggests that larger garage areas are associated with higher sale prices.")
```
