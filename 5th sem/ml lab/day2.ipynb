```python
import numpy as np

x = np.array([183, 175, 134, 170, 144, 183, 167, 114, 125, 187])
y = np.array([433, 393, 270, 364, 346, 399, 360, 361, 319, 376])

n = len(x)
sum_x = np.sum(x)
sum_y = np.sum(y)
sum_xy = np.sum(x * y)
sum_x2 = np.sum(x ** 2)

m = (n * sum_xy - sum_x * sum_y) / (n * sum_x2 - sum_x ** 2)
b = (sum_y - m * sum_x) / n

print(f"Slope (m): {m:.4f}")
print(f"Intercept (b): {b:.4f}")
#print(f"Linear Equation: y = {m:.4f}x + {b:.4f}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([183, 175, 134, 170, 144, 183, 167, 114, 125, 187])
y = np.array([433, 393, 270, 364, 346, 399, 360, 361, 319, 376])

n = len(x)
sum_x = np.sum(x)
sum_y = np.sum(y)
sum_xy = np.sum(x * y)
sum_x2 = np.sum(x ** 2)

m = (n * sum_xy - sum_x * sum_y) / (n * sum_x2 - sum_x ** 2)
b = (sum_y - m * sum_x) / n

x_line = np.linspace(min(x) - 10, max(x) + 10, 100)
y_line = m * x_line + b

plt.figure(figsize=(8, 6))
plt.scatter(x, y, color='blue', label='Actual Data')
plt.plot(x_line, y_line, color='red', label=f'Regression Line: y = {m:.2f}x + {b:.2f}')
plt.title("Problem 2: Regression Model Plot")
plt.xlabel("College Test Scores")
plt.ylabel("University Test Scores")
plt.grid(True)
plt.legend()
plt.show()
```

```python
import numpy as np

x = np.array([183, 175, 134, 170, 144, 183, 167, 114, 125, 187])
y = np.array([433, 393, 270, 364, 346, 399, 360, 361, 319, 376])

n = len(x)
sum_x = np.sum(x)
sum_y = np.sum(y)
sum_xy = np.sum(x * y)
sum_x2 = np.sum(x ** 2)

m = (n * sum_xy - sum_x * sum_y) / (n * sum_x2 - sum_x ** 2)
b = (sum_y - m * sum_x) / n

predicted_y = m * x + b
mse = np.mean((y - predicted_y) ** 2)

print("Problem 3: Model Evaluation")
print("Predicted University Scores (fp(x)):")
print(np.round(predicted_y, 2))
print(f"Mean Squared Error (MSE): {mse:.4f}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([183, 175, 134, 170, 144, 183, 167, 114, 125, 187])
y = np.array([433, 393, 270, 364, 346, 399, 360, 361, 319, 376])

n = len(x)
sum_x = np.sum(x)
sum_y = np.sum(y)
sum_xy = np.sum(x * y)
sum_x2 = np.sum(x ** 2)

m = (n * sum_xy - sum_x * sum_y) / (n * sum_x2 - sum_x ** 2)
b = (sum_y - m * sum_x) / n

predicted_y = m * x + b

x_new = 150
y_new = m * x_new + b

plt.figure(figsize=(8, 6))
plt.scatter(x, y, color='blue', label='Actual Data')
x_line = np.linspace(min(x)-10, max(x)+10, 100)
y_line = m * x_line + b
plt.plot(x_line, y_line, color='red', label='Regression Line')
plt.scatter(x, predicted_y, color='orange', label='Predicted fp(x)', marker='x')
plt.scatter(x_new, y_new, color='green', s=100, label=f'Prediction at x=150: y={y_new:.2f}', edgecolors='black')
plt.title("Problem 4: Regression with Prediction at x = 150")
plt.xlabel("College Test Scores")
plt.ylabel("University Test Scores")
plt.legend()
plt.grid(True)
plt.show()

print(f"Predicted university score for college score 150: {y_new:.2f}")
```
